## useful functions

```python
import pickle
def save_item(p, path):
    with open(path, 'wb') as f:
        pickle.dump(p, f, pickle.HIGHEST_PROTOCOL)
    print("%s: Saved at %s"%(time.ctime(),path))
          
def load_item(path):
    with open(path, 'rb') as f:
        return pickle.load(f)
    print("%s: Load from %s"%(time.ctime(), path))
```

```python
params = {}

params['gpu_device'] = 1

params['batch_size'] = 8
params['max_iteration'] = 100000
params['check_point'] = 2000

params['generator_steps'] = 1
params['generator_learning_rate'] = 2e-5
params['g_fake_weight'] = 1
params['l1_weight'] = 1

params['discriminator_steps'] = 1
params['discriminator_learning_rate'] = 2e-5
params['d_fake_weight'] = 1
params['d_real_weight'] = 1

params['beta1'] = 0.5
params['beta2'] = 0.999

params['imsize'] = 512
params['cropsize'] = 512

params['save_path'] = '../trained_models/cgan/%s/'%(time.strftime('%y-%m-%d_%H-%M-%S'))

######### dependent paramsters
os.environ['CUDA_VISIBLE_DEVICES'] = str(params['gpu_device'])
device = torch.device('cuda') if params['gpu_device'] >= 0 else torch.device('cpu')

if not os.path.exists(params['save_path']):
    os.makedirs(params['save_path'])
    os.makedirs(params['save_path']+'logs/')

save_item(params, params['save_path']+'parameters.pkl')
```

## dataset

```python
denormalize = transforms.Normalize((-0.5, -0.5, -0.5), (1/0.5, 1/0.5, 1/0.5))
unloader = transforms.ToPILImage()

def imshow(x, title):
    x = x.cpu()
    x = denormalize(x)
    x.clamp_(0.0, 1.0)
    
    plt.figure(figsize=(7,7))
    if title:
        plt.title(title)
    plt.imshow(unloader(x))
    plt.show()
    
def imsave(x, path):
    x = x.cpu()
    x = denormalize(x)
    x.clamp_(0.0, 1.0)    
    torchvision.utils.save_image(x,path)
    
    
class FacadeDataset(torch.utils.data.Dataset):

    def __init__(self, dataset_path, input_filelist, target_filelist, imsize=512, cropsize=512):
        
        self.imsize = imsize
        self.cropsize = cropsize
        
        self.resize_ = transforms.Resize(self.imsize)
        self.normalize = transforms.Normalize( (0.5,0.5,0.5), (0.5,0.5,0.5) )
        self.totensor_ = transforms.ToTensor()
           
        
        self.dataset_path = dataset_path
       
        self.input_names = [name.strip() for name in open(input_filelist, 'r').readlines()]
        self.target_names = [name.strip() for name in open(target_filelist, 'r').readlines()]
        assert len(self.input_names) == len(self.target_names)
        

    def __len__(self):
        return len(self.input_names)

    def __getitem__(self, idx):
        input_name = os.path.join(self.dataset_path, self.input_names[idx])
        target_name = os.path.join(self.dataset_path, self.target_names[idx])
        
        input = Image.open(input_name).convert("RGB")
        target = Image.open(target_name).convert("RGB")
        
        # resize
        input = self.resize_(input)
        target = self.resize_(target)
        
        # random crop
        input, target = self.RandomCrop(input, target)
        
        # totensor
        input = self.normalize(self.totensor_(input))
        target = self.normalize(self.totensor_(target))
        
        return input, target
        
    def RandomCrop(self, input, target):
        i, j, h, w = transforms.RandomCrop.get_params(input, output_size=(self.cropsize, self.cropsize))
        input = transforms.functional.crop(input, i, j, h, w)
        target= transforms.functional.crop(target, i, j, h, w)                
        return input, target
        
```

## network

```python
class Network(nn.Module):
    def __init__(self, in_dim, paddings, convs, normalizations, dropouts, activations):
        super(Network, self).__init__()
        
        # network configurations
        self.paddings = paddings
        self.convs = convs
        self.normalizations = normalizations
        self.dropouts = dropouts
        self.activations = activations
        
        # network
        self.network = nn.Sequential()
        
        # make network layers
        in_ch = in_dim
        count = 0
        for padding, conv, norm, dropout, activation in zip(self.paddings, self.convs, self.normalizations, self.dropouts, self.activations):                
            # padding
            if padding:
                padding_size = 1
                if padding == 'zero':
                    self.network.add_module(str(count), nn.ZeroPad2d(padding_size))
                elif padding == 'reflection':
                    self.network.add_module(str(count), nn.ReflectionPad2d(padding_size))
                else:
                    raise RuntimeError("padding flag error")
                count += 1                    
            
            # convolution
            if conv:
                if 'c' in conv:
                    out_ch = int(conv.split('-')[1])
                    self.network.add_module(str(count), nn.Conv2d(in_ch, out_ch, kernel_size=4, stride=2))
                    in_ch = out_ch
                    
                elif 't' in conv:
                    out_ch = int(conv.split('-')[1])
                    self.network.add_module(str(count), nn.ConvTranspose2d(in_ch, out_ch, kernel_size=4, stride=2, padding=1))
                    in_ch = out_ch
                    
                elif 's' in conv:
                    in_ch = int(conv.split('-')[1])
                    out_ch = in_ch//2
                    self.network.add_module(str(count), nn.ConvTranspose2d(in_ch, out_ch, kernel_size=4, stride=2, padding=1))
                    in_ch = out_ch*2
                    
                elif 'h' in conv:
                    in_ch = int(conv.split('-')[1])
                    out_ch = in_ch//4
                    self.network.add_module(str(count), nn.ConvTranspose2d(in_ch, out_ch, kernel_size=4, stride=2, padding=1))
                    in_ch = out_ch*2
                    
                elif 'o' in conv:
                    out_ch = int(conv.split('-')[1])
                    self.network.add_module(str(count), nn.ConvTranspose2d(in_ch, out_ch, kernel_size=4, stride=2, padding=1))
            
                count += 1
                
            
            # normalization
            if norm:
                if norm == 'bn':
                    self.network.add_module(str(count), nn.BatchNorm2d(out_ch))
                    count += 1
                elif norm == 'in':
                    self.network.add_module(str(count), nn.InstanceNorm2d(out_ch))
                    coutn +=1
                else:
                    raise RuntimeError('normalization flag error')
            
            # dropout
            if dropout:
                self.network.add_module(str(count), nn.Dropout2d(p=dropout, inplace=True))
                count += 1
            
            
            # activation
            if activation:
                if activation.lower() == 'relu':
                    self.network.add_module(str(count), nn.ReLU())
                    count += 1
                elif activation.lower() == 'tanh':
                    self.network.add_module(str(count),nn.Tanh())
                    count += 1
                elif activation.lower() == 'sigmoid':
                    self.network.add_module(str(count),nn.Sigmoid())
                    count += 1
                elif 'lrelu' in activation :
                    rate = float(activation.split('-')[-1])
                    self.network.add_module(str(count),nn.LeakyReLU(rate))
                    count += 1
                else:
                    raise RuntimeError("activation flag error")                
            
    def forward(self, x, extract_layer_indices=None, skip_features=None):
        
        # extract features for skip-connected decoder networks
        if extract_layer_indices:
            features = []
            for index, layer in enumerate(self.network):
                x = layer(x)
                if index in extract_layer_indices:
                    features.append(x)
            return x, features
        
        # decoder forwarding with skip-connected encoder features
        elif skip_features:
            for layer in self.network:
                if skip_features and isinstance(layer, nn.ReLU):
                    x = layer(x)
                    skip_connected_feature = skip_features.pop()
                    x = torch.cat([x, skip_connected_feature], 1)
                else:
                    x = layer(x)
            return x
        
        # forwarding
        else:                
            return self.network(x)
```

```python
class Generator(nn.Module):
    def __init__(self, encoder_configurations, decoder_configurations, extract_layer_index):
        super(Generator, self).__init__()
        self.extract_layer_index = extract_layer_index
        
        self.encoder = Network(in_dim = encoder_configurations['in_dim'],
                              paddings = encoder_configurations['paddings'],
                              convs = encoder_configurations['convolutions'],
                              normalizations = encoder_configurations['normalizations'],
                              dropouts = encoder_configurations['dropouts'],
                              activations = encoder_configurations['activations'])
        
        self.decoder = Network(in_dim = decoder_configurations['in_dim'],
                              paddings = decoder_configurations['paddings'],
                              convs = decoder_configurations['convolutions'],
                              normalizations = decoder_configurations['normalizations'],
                              dropouts = decoder_configurations['dropouts'],
                              activations = decoder_configurations['activations'])
        
    def forward(self, x):
        z, skip_features = self.encoder(x, extract_layer_indices = self.extract_layer_index)
        return self.decoder(z, skip_features=skip_features)
```

```python
class Discriminator(nn.Module):
    def __init__(self, configurations):
        super(Discriminator, self).__init__()
        self.network = Network( in_dim = configurations['in_dim'], 
                               paddings = configurations['paddings'], 
                               convs = configurations['convolutions'], 
                               normalizations = configurations['normalizations'], 
                               dropouts = configurations['dropouts'],
                               activations = configurations['activations'],
                              )
    def forward(self, x):
        return self.network(x)
```

```python
# network configurations from the paper.
encoder_info = {}
encoder_info['in_dim'] = 3
encoder_info['paddings'] = ['zero', 'zero','zero','zero','zero','zero','zero','zero']
encoder_info['convolutions'] = ['c-64', 'c-128', 'c-256', 'c-512', 'c-512', 'c-512', 'c-512', 'c-512']
encoder_info['normalizations'] = [None, 'bn', 'bn', 'bn', 'bn', 'bn', 'bn', 'bn']
encoder_info['dropouts'] = [None, None, None, None, None, None, None, None]
encoder_info['activations'] = ['lrelu-0.2', 'lrelu-0.2','lrelu-0.2','lrelu-0.2','lrelu-0.2','lrelu-0.2','lrelu-0.2','lrelu-0.2']

decoder_info = {}
decoder_info['in_dim'] = 512
decoder_info['paddings'] = [None, None, None, None, None, None, None, None]
decoder_info['convolutions'] = ['t-512', 's-1024', 's-1024', 's-1024', 'h-1024', 'h-512', 'h-256', 'o-3']
decoder_info['normalizations'] = [None, 'bn', 'bn', 'bn', 'bn', 'bn', 'bn', None]
decoder_info['dropouts'] = [0.5, 0.5, 0.5, None, None, None, None, None]
decoder_info['activations'] = ['relu', 'relu','relu','relu','relu','relu','relu', 'tanh']

discriminator_info = {}
discriminator_info['in_dim'] = 6
discriminator_info['paddings'] = ['zero', 'zero', 'zero', 'zero', 'zero']
discriminator_info['convolutions'] = ['c-64', 'c-128', 'c-256', 'c-512', 'c-1']
discriminator_info['normalizations'] = [None, 'bn', 'bn', 'bn', None]
discriminator_info['dropouts'] = [None, 0.5, 0.5, 0.5, None]
discriminator_info['activations'] = ['lrelu-0.2', 'lrelu-0.2', 'lrelu-0.2', 'lrelu-0.2', 'sigmoid']

#
network_configurations = {}
network_configurations['extract_encoder_layers'] = [2, 6, 10, 14, 18, 22, 26]

network_configurations['encoder_network_configuration'] = encoder_info
network_configurations['decoder_network_configuration'] = decoder_info
network_configurations['discriminator_network_configuration'] = discriminator_info

save_item(network_configurations, params['save_path']+'network_configurations.pkl')
```

### loss and optimizer

```python
bce_criterion = nn.BCELoss()
l1_criterion = nn.L1Loss()
```

```python
def get_optimizer(network, lr, b1, b2):
    return torch.optim.Adam(network.parameters(), lr=lr, betas=(b1,b2))
```

### train

```python
# generator network
generator_network = Generator(encoder_configurations=encoder_info, 
                              decoder_configurations=decoder_info, 
                              extract_layer_index=network_configurations['extract_encoder_layers']).to(device)

# discriminator network
discriminator_network = Discriminator(discriminator_info).to(device)
```

```python
dataset_path = '../dataset/cmp-facade-dataset/base/'
input_filelist = '../dataset/cmp-facade-dataset/base-mask.txt'
target_filelist = '../dataset/cmp-facade-dataset/base-image.txt'

# dataset
dataset = FacadeDataset(dataset_path, input_filelist, target_filelist,imsize=params['imsize'], cropsize=params['cropsize'])
```

```python
# optimizer
g_optimizer = get_optimizer(generator_network, 
                            lr=params['generator_learning_rate'], 
                            b1=params['beta1'], b2=params['beta2'])
d_optimizer = get_optimizer(discriminator_network, 
                            lr=params['discriminator_learning_rate'], 
                            b1=params['beta1'], b2=params['beta2'])
```

```python
loss_log = {}
loss_log['g_fake_loss'] = []
loss_log['l1_loss'] = []

loss_log['d_fake_loss'] = []
loss_log['d_real_loss'] = []

for iteration in range(params['max_iteration']+1):
    g_loss_ = 0
    g_l1_loss_ = 0
    for k in range(params['generator_steps']):        
        # train generator network
        label, image = next(iter(torch.utils.data.DataLoader(dataset, batch_size=params['batch_size'],shuffle=True)))
        label, image = label.to(device), image.to(device)

        fake = generator_network(label)

        fake_inference = discriminator_network(torch.cat([label, fake], 1))

        fake_loss = bce_criterion(fake_inference, torch.ones_like(fake_inference).to(device))
        l1_loss = l1_criterion(fake, image)

        generator_loss = fake_loss*params['g_fake_weight'] + l1_loss * params['l1_weight']    

        g_optimizer.zero_grad()
        generator_loss.backward()
        g_optimizer.step()
        
        g_loss_ += fake_loss.item()
        g_l1_loss_ += l1_loss.item()
    
    # log generator losses
    loss_log['g_fake_loss'].append(g_loss_/params['generator_steps'])
    loss_log['l1_loss'].append(g_l1_loss_/params['generator_steps'])
    
    
    
    # train discriminator network
    d_fake_loss_ = 0
    d_real_loss_ = 0
    for k in range(params['discriminator_steps']):
        label, image = next(iter(torch.utils.data.DataLoader(dataset, batch_size=params['batch_size'],shuffle=True)))
        label, image = label.to(device), image.to(device)

        fake = generator_network(label)
        fake_inference = discriminator_network(torch.cat([label, fake], 1))    
        fake_loss = bce_criterion(fake_inference, torch.zeros_like(fake_inference).to(device))

        real_inference = discriminator_network(torch.cat([label, image], 1))    
        real_loss =bce_criterion(real_inference, torch.ones_like(real_inference).to(device))

        discriminator_loss = fake_loss * params['d_fake_weight'] + real_loss * params['d_real_weight']

        d_optimizer.zero_grad()
        discriminator_loss.backward()
        d_optimizer.step()
        
        d_fake_loss_ += fake_loss.item()
        d_real_loss_ += real_loss.item()
    
    # log discriminator losses
    loss_log['d_fake_loss'].append(d_fake_loss_/params['discriminator_steps'])
    loss_log['d_real_loss'].append(d_real_loss_/params['discriminator_steps'])
    
    if iteration%params['check_point'] == 0:
        log_str = '%s: '%(time.strftime('%y-%m-%d %H:%M:%S'))
        log_str += ', generator L1 loss : %2.4f'%(np.mean(loss_log['l1_loss']))
        log_str += ', generator loss : %2.4f'%(np.mean(loss_log['g_fake_loss']))
        log_str += ', discriminator fake loss : %2.4f'%(np.mean(loss_log['d_fake_loss']))
        log_str += ', discriminator real loss : %2.4f'%(np.mean(loss_log['d_real_loss']))
        print(log_str)
        
        # image show
        input_label_map = torchvision.utils.make_grid(label, nrow=params['batch_size']//2, padding=10, pad_value=1)
        ground_truth_image = torchvision.utils.make_grid(image, nrow=params['batch_size']//2, padding=10, pad_value=1)
        generator_output_image = torchvision.utils.make_grid(fake, nrow=params['batch_size']//2, padding=10, pad_value=1)
        
        imshow(input_label_map, "Input Label Map")
        imshow(ground_truth_image, "Ground Truth")        
        imshow(generator_output_image, "Generator's output")
                
        # loss plot
        for key, value in loss_log.items():
            plt.semilogy(value, label=key)
        plt.legend()
        plt.show()
    
        imsave(input_label_map, params['save_path']+'logs/iteration-%d_input-label-map.jpg'%iteration)
        imsave(ground_truth_image, params['save_path']+'logs/iteration-%d_ground-truth-image.jpg'%iteration)
        imsave(generator_output_image, params['save_path']+'logs/iteration-%d_generator-output.jpg'%iteration)        
        save_item(loss_log, params['save_path']+'loss_logs.pkl')        
        torch.save(generator_network.state_dict(), params['save_path']+'generator.model')
        torch.save(discriminator_network.state_dict(), params['save_path']+'discriminator.model')
```
